This notebook creates the heat map for "Figure 7: Effect of Mangroves Distance and Width on Sale Price, Medium Risk Areas"

```python
import numpy as np
import matplotlib.pyplot as plt

# Regression coefficients
coefficient_distance = -0.025 
# source of -0.025: 2nd year effect for on the path Hurricaneafter * Mangrove * Medium inTable 2: Effect of Mangroves on Prices 
# after Hurricanes: Accounting for Hurricane Path)
coefficient_width = 0.08
# using positive value here because higher width means more mangrove protection
# source of -0.08: 2nd year effect for major hurricane in table_width_inv_newbenchmark.tex (\label{t.w.inv.2.mid} 

#Use Hurricane Ivan of year 2
# Generating sample data for distance and width
# distance is in km and width is in meter
distance_values = np.linspace(0, 10, 50)  # Replace 0 and 10 with your desired range
width_values = np.linspace(5, 0, 80)     # Replace 0 and 5 with your desired range

# Creating a meshgrid of distance and width values
distance_mesh, width_mesh = np.meshgrid(distance_values, width_values)

# Calculating effects based on the coefficients
effects = (
    coefficient_distance * distance_mesh +
    coefficient_width * width_mesh 
    # coefficient_interaction * distance_mesh * width_mesh
)

# Plotting the heatmap
# plt.figure(figsize=(8, 6))
# heatmap = plt.pcolormesh(distance_values, width_values, effects, cmap='viridis')
# plt.colorbar(heatmap, label='Effect')
# plt.xlabel('Distance')
# plt.ylabel('Width')
# plt.title('Effect Based on Regression Coefficients')
# plt.show()
# Plotting the heatmap with gridlines
plt.figure(figsize=(6.5, 5))
heatmap = plt.pcolormesh(distance_values, width_values, effects, cmap='magma')

# Adding gridlines
plt.grid(visible=True, linestyle='--', linewidth=0.5, color='black', alpha=0.5)

# Adjusting the colorbar label size and weight
cbar = plt.colorbar(heatmap)
cbar.set_label('Sale Price Change', weight='bold', size='large')

# Adjusting the xlabel and ylabel size and weight
plt.xlabel('Distance (km)', weight='bold', size='large')
plt.ylabel('Width (10m)', weight='bold', size='large')

plt.savefig('./Analysis/Results/Figures/distance_width_heatmap_newbenchmark_path.png', dpi=500, bbox_inches='tight')
plt.show()
```
